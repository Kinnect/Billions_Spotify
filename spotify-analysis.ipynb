```python
top_songs.rename(columns={"Track Name": "name"},inplace=True)
top_songs
```

```python
song_info.rename(columns={"release_date": "Album Release Date"},inplace=True)
song_info.drop('Album Release Date', axis=1, inplace=True)
song_info
```

```python
song_list = pd.merge(top_songs,song_info,on=['name'])
song_list
```

```python
song_list_f = song_list.sort_values('popularity', ascending=False).drop_duplicates('name').sort_index()
song_list_f.to_csv('test.csv')
```

```python
columns_to_drop = ['Track URI', 'Artist URI(s)','Artist Name(s)', 'Album URI', 'Album Name', 'Album Artist URI(s)', 'Album Artist Name(s)','Album Image URL',
    'Disc Number','Track Preview URL','Explicit','Added By','Added At','id','popularity','duration_ms','artists','id_artists']
```

```python
song_list_f = song_list_f.drop(columns_to_drop, axis = 1)
song_list_f
```

```python
song_list_t = song_list_f.drop(['name','Album Release Date'], axis=1)
song_list_t.index = np.arange(1, len(song_list_t) + 1)
song_list_t.reset_index(inplace=True)
song_list_t = song_list_t.rename(columns = {'index':'Rank'})
song_list_t
```

```python
plt.figure(figsize=(14,6))
sns.regplot(x=song_list_t['Rank'], y=song_list_t['Track Duration (ms)'])
```
